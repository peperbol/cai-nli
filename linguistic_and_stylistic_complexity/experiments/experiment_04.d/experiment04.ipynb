```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import scipy
import seaborn as sns

plt.rcParams['figure.figsize'] = (12.0, 8.0)
```

```python
data = pd.read_csv("pos_distribution.tsv", sep="\t", header=0, index_col=0)
```

```python
genres = "high romance scifi horror".split()
default_palette = sns.color_palette()
new_palette = [default_palette[0]] + default_palette[2:]
sns.set_palette(new_palette)

sns.countplot(x="genre", data=data)
plt.rcParams['figure.figsize'] = (12.0, 8.0)
```

```python
def plot_tags(tags):
    tgs = tags.split()
    if len(tgs) > 1:
        d = pd.DataFrame()
        d["total"] = data[tgs].sum(axis=1)
        d["genre"] = data["genre"]
        plt.figure()
        sns.boxplot(x="genre", y="total", data=d)
    for tag in tgs:
        plt.figure()
        sns.boxplot(x="genre", y=tag, data=data)
```
